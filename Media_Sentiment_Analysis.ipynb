```python
plt.scatter(sentiments_cbs["Tweets Ago"], sentiments_cbs["Compound"], marker="o", c ="green", edgecolors="black",
            label='CBS', alpha=.7)
plt.scatter(sentiments_fox["Tweets Ago"], sentiments_fox["Compound"], marker="o", c ="blue", edgecolors="black",
            label='FOX', alpha=.7)
plt.scatter(sentiments_cnn["Tweets Ago"], sentiments_cnn["Compound"], marker="o", c ="red", edgecolors="black",
            label='CNN', alpha=.7)
plt.scatter(sentiments_bbc["Tweets Ago"], sentiments_bbc["Compound"], marker="o", c ="skyblue", edgecolors="black",
            label='BBC', alpha=.7)
plt.scatter(sentiments_times["Tweets Ago"], sentiments_times["Compound"], marker="o", c ="gold", edgecolors="black",
            label='New York Times', alpha=.7)
plt.title("Sentiment Analysis of Media Tweets (12/21/17)")
plt.grid()
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.savefig("Sentiment_Scatter.png")
#legend = plt.legend(loc='best', shadow=True)
plt.show()
```

```python
#Get Overall Sentiment Values per Media
overall_cbs = sentiments_cbs["Compound"].mean()
overall_fox = sentiments_fox["Compound"].mean()
overall_cnn = sentiments_cnn["Compound"].mean()
overall_bbc = sentiments_bbc["Compound"].mean()
overall_times = sentiments_times["Compound"].mean()
```

```python
#Plot Bar Graph Showing Overall Sentiment
compound_sentiments =[overall_cbs, overall_fox, overall_cnn, overall_bbc, overall_times]
colors=['green','blue','red','skyblue','gold']
x_axis = np.arange(len(compound_sentiments))
plt.bar(x_axis, compound_sentiments, color=colors, alpha=0.5, align="edge", width= 1.0)
tick_locations = [value+0.5 for value in x_axis]
plt.xticks(tick_locations, ["CBS", "FOX", "CNN", "BBC", "New York Times"])
plt.title("Overall Media Sentiment based on Twitter (12/21)")
plt.ylabel("Tweet Polarity")
plt.xlim(0, len(x_axis))
plt.grid()
plt.savefig("Sentiment_Overall_Bar.png")
plt.show()
```

Obserations:
1. BBC covers a variety of information compared to the other media outlets. You can make an assumption that its diversity of journalism away from typical political news, is why BBC has the highest Polarity score. 

2. Majority of the tweets sit at a neutral level. You can make an assumption that media outlets overall goal is to remain neutral in its media coverage and report facts. 

3. The TV media outlets have the lowest Polarity scores, with the 24/7 cable shows, CNN & Fox, have the lowest. The trend is media coverage on tv tends to be more negative.
